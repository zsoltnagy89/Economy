## Notebook to visualize new datatables

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load data
path_data = '../input/TSN_Q_exp.csv'
data = pd.read_csv(path_data, sep=';', parse_dates=[2])
# filter data
report = 'Q'
start_date = pd.Timestamp('2001-01-01')
data_filt = data.loc[(data['report'] == report) & (data['date'] > start_date), :]

data_filt.reset_index(drop=True, inplace=True)
```

```python
for column in data_filt.columns[3:]:
    data_filt[column] = data_filt[column].astype('float')
```

```python
data_filt.head()
```

```python
for column in data_filt.columns[3:]:
        fig, ax1 = plt.subplots(figsize=(15, 6))

        ax1.plot(data_filt['date'], data_filt[column], color='k', label=column)

        ax1.set_xlabel('Date')
        ax1.set_ylabel(column.upper())
        
        plt.legend()
        plt.show()
```
